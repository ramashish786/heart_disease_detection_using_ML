```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Reading raw data

df_raw = pd.read_csv("Data/heart_2022_with_nans.csv")
df_raw.head(5)
```

```python
# Dropping null values from dataframe
df_nn = df_raw.dropna()
```

```python
df_nn.describe()
```

```python
for col in df_nn.columns:
    print(col)
    print(df_nn[col].value_counts())
```

```python
# dropping unwanted columns

df_cd = df_nn.drop(['State', 'DifficultyWalking','DifficultyConcentrating','BlindOrVisionDifficulty','DifficultyDressingBathing','ECigaretteUsage','ChestScan','RaceEthnicityCategory','HIVTesting','PneumoVaxEver','HighRiskLastYear','CovidPos'],axis=1)
```

```python
df_cd.shape
```

```python
df_cd.info()
```

```python
YES_NO_QUESTIONS = {'Yes':1,'No':0}

df_cd['HadHeartAttack'] = df_cd['HadHeartAttack'].map(YES_NO_QUESTIONS)
```

```python
df_cd['HadHeartAttack'].dtype
```

```python
categorical_columns = df_cd.select_dtypes(include=['object']).columns.tolist()
```

```python
categorical_columns
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(handle_unknown = 'ignore')
```

```python
one_hot_encoded = encoder.fit_transform(df_cd[categorical_columns])
```

```python
one_hot_encoded.shape
```

```python
np.array(encoder.get_feature_names_out(categorical_columns)).shape
```

```python
one_hot_df = pd.DataFrame(one_hot_encoded)
```

```python
df_encoded = pd.concat([df_cd, one_hot_df], axis=1)
```
